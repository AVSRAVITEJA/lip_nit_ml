```python
df= pd.read_csv("heart-disease.csv")
df.shape
```

```python
sex=np.unique(df["sex"])
print(sex)
```

```python
fbs=np.unique(df["fbs"])
print(fbs)
```

```python
pp=[]
point=0
for i in sex:
    pp.append([])
    for j in df["sex"]:
        if j==i:
            pp[point].append(1)
        else:
            pp[point].append(0)
    point+=1
print(pp)

for i in range(len(sex)):
    df[sex[i]] = pp[i]
```
